```python
# Install necessary libraries
!pip install yfinance pandas

import yfinance as yf
import pandas as pd

# Function to fetch stock data and calculate various indicators
def calculate_technical_indicators_and_pe_rsi(stock_symbol):
    # Fetch historical stock data for 90 months
    stock_data = yf.download(stock_symbol, period="10mo")
    
    # Calculate the 20-day simple moving average
    stock_data['20_Day_SMA'] = stock_data['Close'].rolling(window=20).mean()
    
    # Calculate the rolling standard deviation
    stock_data['20_Day_STD'] = stock_data['Close'].rolling(window=20).std()
    
    # Calculate the upper and lower Bollinger Bands
    stock_data['Upper_Band'] = stock_data['20_Day_SMA'] + (stock_data['20_Day_STD'] * 2)
    stock_data['Lower_Band'] = stock_data['20_Day_SMA'] - (stock_data['20_Day_STD'] * 2)
    
    # Calculate the True Range (TR)
    stock_data['High-Low'] = stock_data['High'] - stock_data['Low']
    stock_data['High-PrevClose'] = abs(stock_data['High'] - stock_data['Close'].shift(1))
    stock_data['Low-PrevClose'] = abs(stock_data['Low'] - stock_data['Close'].shift(1))
    stock_data['TR'] = stock_data[['High-Low', 'High-PrevClose', 'Low-PrevClose']].max(axis=1)
    
    # Calculate the Average True Range (ATR)
    stock_data['ATR'] = stock_data['TR'].rolling(window=14).mean()
    
    # Drop the intermediate columns used for ATR calculation
    stock_data.drop(['High-Low', 'High-PrevClose', 'Low-PrevClose'], axis=1, inplace=True)
    
    # Calculate the Relative Strength Index (RSI)
    delta = stock_data['Close'].diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=14).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=14).mean()
    rs = gain / loss
    stock_data['RSI'] = 100 - (100 / (1 + rs))
    
    stock_data['Pct_Change'] = stock_data['Close'].pct_change() * 100

    bitcoin_data = yf.download('BTC-USD', period="10mo")
    
    # Calculate the percentage change in Bitcoin price
    bitcoin_data['Pct_Change_BTC'] = bitcoin_data['Close'].pct_change() * 100
    
    # Align Bitcoin percentage change data with stock data based on dates
    stock_data['Pct_Change_BTC'] = bitcoin_data['Pct_Change_BTC']
    
    def classify_change(x):
        if x <= 0:
            return 0
        elif x > 0:
            return 1
        else:
            return 0
    
    stock_data['Pct_Change_Class'] = stock_data['Pct_Change'].apply(classify_change)
    stock_data['Pct_Change_BTC_Class'] = stock_data['Pct_Change_BTC'].apply(classify_change)
    
    stock_data.dropna(inplace=True)
    
    return stock_data

# Example usage
stock_symbol = 'COIN'  # Example stock symbol for Apple Inc.
stock_data = calculate_technical_indicators_and_pe_rsi(stock_symbol)

# Display the last few rows of the DataFrame
print(stock_data.tail())

```

```python
from sklearn.model_selection import train_test_split
```

```python
X = stock_data[['RSI']]
y = stock_data['Pct_Change_Class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
logmodel = LogisticRegression(max_iter=1000)
logmodel.fit(X_train,y_train)
```

```python
preds = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,preds))
```
